# **Importing Dataset**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import regression
sns.set()
plt.style.use('seaborn-whitegrid')

data = pd.read_csv('https://raw.githubusercontent.com/nsantivardhan/DogeCoin-Regression-model-/main/DOGE-USD.csv')
print(data.head())
```

## **Plotting the DATASET**

```python
data.dropna()
plt.figure(figsize=(10, 4))
plt.title("DogeCoin Price USD")
plt.xlabel("Date")
plt.ylabel("Close")
plt.plot(data["Close"])
plt.show()
```

### **Using Regression to predict the future closing value**

```python
!pip install autots
from autots import AutoTS
model = AutoTS(forecast_length=10, frequency='infer', ensemble='simple', drop_data_older_than_periods=200)
model = model.fit(data, date_col='Date', value_col='Close', id_col=None)

prediction = model.predict()
forecast = prediction.forecast
print("DogeCoin Price Prediction")
print(forecast)
```
